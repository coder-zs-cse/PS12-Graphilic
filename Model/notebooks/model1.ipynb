```python
with open(r'C:\\Users\\91932\\Desktop\\hackrx4.0\\PS12-Graphilic\\Data\\user_user_graph.json','rb') as f:
    data = json.load(f)
graph = nx.node_link_graph(data)
```

```python
dataset = pd.read_csv("C:\\Users\\91932\\Desktop\\hackrx4.0\\PS12-Graphilic\\Data\\user-user-dataframe.csv")
```

```python
def get_similarity_list(model,customerID,list_items):
    cust_embed = model.wv[customerID]
    final_items = {}

    for i in list_items:
        i = re.findall(r'\d+', i)
        i = ''.join(i)
        try:
            item_embedding = model.wv[i]
            final_items[i] = cust_embed.dot(item_embedding)
        except KeyError:
            continue
    return final_items
```

```python
'''find all the neighbouring users of a particular user'''
custid = 'A2JW67OY8U6HHK'

neighbours = list(graph.neighbors(custid))
list_items = []
for node in neighbours:
    list_items.append(dataset.loc[dataset['customer id']==node,'ASIN'].iloc[0])
```

```python
prediction = get_similarity_list(model,'A2JW67OY8U6HHK',list_items)
```
